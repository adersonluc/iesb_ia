```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

```python
temp = ctrl.Antecedent(np.arange(0, 50, 1), 'temp')
sol = ctrl.Antecedent(np.arange(0,100,5), 'sol')
ocupacao = ctrl.Antecedent(np.arange(0,100,5), 'ocupacao')
```

```python
temp.automf(5)
sol.automf(3)
```

```python
ocupacao['baixa'] = fuzz.trimf(ocupacao.universe, [0, 0, 13])
ocupacao['media'] = fuzz.trimf(ocupacao.universe, [0,13,25])
ocupacao['alta'] = fuzz.trimf(ocupacao.universe, [13, 25, 25])
```
